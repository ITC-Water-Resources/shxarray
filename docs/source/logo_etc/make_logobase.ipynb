```python
import matplotlib.pyplot as plt
import matplotlib.colors as cm
# from matplotlib.transforms import Bbox
import numpy as np
```

```python
%matplotlib widget
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection='3d')

div=2
ndiscr=100

# Make data
u = np.linspace(0, 0.5*np.pi, ndiscr)
u2 = np.linspace(1*np.pi, 1.5*np.pi, ndiscr)
v = np.linspace(0, np.pi, ndiscr)

x1 = 10 * np.outer(np.cos(u), np.sin(v))
y1 = 10 * np.outer(np.sin(u), np.sin(v))
z1 = 10 * np.outer(np.ones(np.size(u)), np.cos(v))


x2 = 10 * np.outer(np.cos(u2), np.sin(v))
y2 = 10 * np.outer(np.sin(u2), np.sin(v))
z2 = 10 * np.outer(np.ones(np.size(u2)), np.cos(v))

# Plot the surfaces
ls = cm.LightSource(-100, 0)
# To use a custom hillshading mode, override the built-in shading and pass
# in the rgb colors of the shaded surface calculated from "shade".
# rgb = ls.shade(z, cmap=cm.gist_earth, vert_exag=0.1, blend_mode='soft')

# logocolor='#007fc3'
logocolor1='#40c4ff'
logocolor2='#007fc3'

shade=False
alpha=0.5
ax.plot_surface(x1, y1, z1,color=logocolor1,edgecolor=None,shade=shade,lightsource=ls,alpha=alpha)
ax.plot_surface(x2, y2, z2,color=logocolor1,shade=shade,lightsource=ls,alpha=alpha)



# Set an equal aspect ratio
ax.set_aspect('equal')
ax.view_init(elev=-45, azim=135, roll=-45)
plt.axis('off')
plt.tight_layout()
plt.show()
# plt.savefig('logo_base.svg',transparent=True,bbox_inches='tight')
plt.savefig('logo_base.png',transparent=True,dpi=300)
```
